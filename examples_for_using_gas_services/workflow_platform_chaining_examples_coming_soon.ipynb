# Workflow Platform Chaining Examples

This notebook is a placeholder for future examples showing how GAS services can be chained from external workflow platforms and agent orchestration frameworks.

Planned examples will demonstrate service chaining with platforms such as:

- KNIME workflows that call GAS operations through HTTP nodes.
- AutoGen-based orchestration that discovers agents, selects services, executes tasks, and passes artifacts between services.
- Other workflow systems that can consume REST-style JSON APIs.

This example is to show how a workflow platform can coordinate the GAS services by calling `GetCapabilities`, `DescribeAgent`, `ExecuteTask`, `GetTaskStatus`, and `GetTaskResult`, then passing returned artifact URLs or encoded outputs into the next service.
