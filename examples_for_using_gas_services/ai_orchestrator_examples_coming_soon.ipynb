# AI Orchestrator Examples

This notebook is a placeholder for future examples showing how AI orchestrators can discover and call Geospatial Agentic Services (GAS).

Planned examples will demonstrate how an external orchestrator can:

- Read `GetCapabilities` to discover available GAS agents.
- Read `DescribeAgent` documents to understand agent-specific capabilities, input requirements, credentials, task modes, and response artifacts.
- Select one or more GAS agents based on a user goal.
- Execute tasks in `sync`, `async`, or `stream` mode.
- Chain service outputs from one agent into inputs for another agent.
- Track task status and retrieve standard GAS task responses.

The GAS server itself does not coordinate multi-agent collaboration internally. Orchestration is expected to happen outside the server through clients, notebooks, browser applications, workflow systems, or AI orchestrators that consume GAS services through the public interfaces.
